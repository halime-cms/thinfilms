```julia
# Modelleri yukluyorum
using Plots, LaTeXStrings
pyplot()
using ThinFilmsTools
using Optim
using GLM

# Dalga boyu araligi [nm]
λ = 200:2100
# Gelis Acisi [degrees]
θ = [60.]
# Polarizasyon (1.0 = p, 0.0 = s, 0.0 ve 1.0 arasi = ortalama)
pol = 0.5
beam = PlaneWave(λ, θ; p=pol) #ISIN icin duzlem dalga

# Kirilma indeksleri olay (0) ve substrate (2)
incident = RIdb.air(beam.λ)
emergent = RIdb.fused_silica_uv(beam.λ) #ortaya cikan 

# Kullanilacak RI modeli tanimliyoruz
layers = [
    LayerTMMO(incident),
    ModelFit(:tauclorentz),
    LayerTMMO(emergent),
]

# olculen mutlak gecirgenlik
Ψexp, Δexp, = SpectraDB.hafnia_ellips(beam.λ)
plot(beam.λ, Ψexp)
plot!(beam.λ, Δexp)
gui()

# Spektrurum tipi: giris deneysel spekturumlari ile elipsiyometri
spectype = Ellipsometry([Ψexp Δexp])

## BBO kullanarak optimizasyon

# bir osilator
seed = [vcat(100.0, #kalinlik
             [3.0, 5.5,
              545, 6.2, 13.2], # osc 1
        ),
]
lb = 0.01.*seed
ub = 2.0.*seed
lb[1][1] = 80
ub[1][1] = 200
lb[1][3] = 4.0
ub[1][3] = 6.0

solOptim = fit_tmm_optics(
    spectype, seed, beam, layers;
    alg=:BBO,
    SearchRange=Utils.unfold_bnd(lb,ub),
)

plot(FitSpectrumEllip(),
    solOptim.beam.λ, solOptim.spectrumExp, solOptim.spectrumFit,
    xaxis=("Dalga Boyu [nm]"), yaxis=(),
)
gui()
```

```julia
function mse(experimental, modeled)
    # calculate the MSE
    return sum((experimental .- modeled).^2) / length(experimental)
end

# ...

solOptim = fit_tmm_optics(spectype, seed, beam, layers; alg=:BBO,SearchRange=Utils.unfold_bnd(lb,ub),)

# calculate the MSE after the optimization is completed
mse_value = mse(solOptim.spectrumExp, solOptim.spectrumFit)

# print the MSE
println("MSE: $(mse_value)")

# ...
```

```julia
#MSE 0.05 ORTLAMA KARE HATASI (The mean square error )
```

```julia
function mae(experimental, modeled)
    # calculate the MAE
    return sum(abs.(experimental .- modeled)) / length(experimental)
end

# ...

solOptim = fit_tmm_optics(spectype, seed, beam, layers; alg=:BBO,SearchRange=Utils.unfold_bnd(lb,ub),)

# calculate the MAE after the optimization is completed
mae_value = mae(solOptim.spectrumExp, solOptim.spectrumFit)

# print the MAE
println("MAE: $(mae_value)")

# ...
```
